```python
#######
# Side-by-side heatmaps for Houston, New York City and Seattle,
# using a shared temperature scale.
######
import plotly.offline as pyo
import plotly.graph_objs as go
from plotly import tools
import pandas as pd

df1 = pd.read_csv('KHOU.csv')
df2 = pd.read_csv('KNYC.csv')
df3 = pd.read_csv('KSEA.csv')


df1['Year']=df1['date'].str.split('-').str[0]
df1['Month']=df1['date'].str.split('-').str[1]
df1['Day']=df1['date'].str.split('-').str[2]

df2['Year']=df2['date'].str.split('-').str[0]
df2['Month']=df2['date'].str.split('-').str[1]
df2['Day']=df2['date'].str.split('-').str[2]

df3['Year']=df3['date'].str.split('-').str[0]
df3['Month']=df3['date'].str.split('-').str[1]
df3['Day']=df3['date'].str.split('-').str[2]

trace1 = go.Heatmap(
    x=df1['Day'],
    y=df1['Month'],
    z=df1['actual_max_temp'],
    colorscale='Jet',
    zmin = 5, zmax = 115 # add max/min color values to make each plot consistent
)
trace2 = go.Heatmap(
    x=df2['Day'],
    y=df2['Month'],
    z=df2['actual_max_temp'],
    colorscale='Jet',
    zmin = 5, zmax = 115 # add max/min color values to make each plot consistent
)
trace3 = go.Heatmap(
    x=df3['Day'],
    y=df3['Month'],
    z=df3['actual_max_temp'],
    colorscale='Jet',
    zmin = 5, zmax = 115 # add max/min color values to make each plot consistent
)


fig = tools.make_subplots(rows=1, cols=3,
    subplot_titles=('Houston, TX','New York City, NY', 'Seattle, WA'),
    shared_yaxes = False,  # this makes the hours appear only on the left
)
fig.append_trace(trace1, 1, 1)
fig.append_trace(trace2, 1, 2)
fig.append_trace(trace3, 1, 3)

fig['layout'].update(      # access the layout directly!
    title='Daily Max Temperatures,7-1-2015 to 6-30-2015'
)
pyo.iplot(fig)
```

```python
import plotly.express as px
import pandas as pd

data = pd.read_csv('72793524234.csv')

# Convert 'DATE' column to datetime objects
data['DATE'] = pd.to_datetime(data['DATE'])

# Extract date components
data['Year'] = data['DATE'].dt.year
data['Month'] = data['DATE'].dt.month
data['Day'] = data['DATE'].dt.day
data['Hour'] = data['DATE'].dt.hour
data['Minute'] = data['DATE'].dt.minute

# Convert 'HourlyPrecipitation' to numeric, coercing errors to NaN
data['HourlyPrecipitation'] = pd.to_numeric(data['HourlyPrecipitation'], errors='coerce')

# Fill NaN values in 'HourlyPrecipitation' with 0, as missing precipitation data can be assumed as no precipitation
data['HourlyPrecipitation'].fillna(0, inplace=True)

# Group by Month and sum the HourlyPrecipitation
monthly_precipitation = data.groupby('Month')['HourlyPrecipitation'].sum().reset_index()

# Create the bar chart with enhanced features
fig = px.bar(monthly_precipitation, x='Month', y='HourlyPrecipitation',
              title='Total Hourly Precipitation by Month in Seattle',
              labels={'HourlyPrecipitation': 'Total Precipitation (inches)', 'Month': 'Month'},
              color='HourlyPrecipitation', # Color bars by precipitation intensity
              color_continuous_scale='Blues', # Apply a blue color scale
              text='HourlyPrecipitation' # Add text labels to bars
             )

# Update the traces to format text labels to two decimal places and position them outside
fig.update_traces(texttemplate='%{text:.2f}', textposition='outside')

# Display the enhanced figure
fig.show()
print("Enhanced bar chart for Total Hourly Precipitation by Month in Seattle generated.")
```

```python
import pandas as pd

# 1. Load the data into DataFrames
df1 = pd.read_csv('72259299999 (1).csv')
df2 = pd.read_csv('72412003872 (1).csv')
df3 = pd.read_csv('72793524234.csv')

# 2. Convert 'DATE' column to datetime objects
df1['DATE'] = pd.to_datetime(df1['DATE'])
df2['DATE'] = pd.to_datetime(df2['DATE'])
df3['DATE'] = pd.to_datetime(df3['DATE'])

# 3. Extract 'Year', 'Month', 'Day', 'Hour' and 'Minute' into new columns
df1['Year'] = df1['DATE'].dt.year
df1['Month'] = df1['DATE'].dt.month
df1['Day'] = df1['DATE'].dt.day
df1['Hour'] = df1['DATE'].dt.hour
df1['Minute']=df1['DATE'].dt.minute

df2['Year'] = df2['DATE'].dt.year
df2['Month'] = df2['DATE'].dt.month
df2['Day'] = df2['DATE'].dt.day
df2['Hour'] = df2['DATE'].dt.hour
df2['Minute']=df2['DATE'].dt.minute

df3['Year'] = df3['DATE'].dt.year
df3['Month'] = df3['DATE'].dt.month
df3['Day'] = df3['DATE'].dt.day
df3['Hour'] = df3['DATE'].dt.hour
df3['Minute']=df3['DATE'].dt.minute

print("DataFrames prepared with 'Year', 'Month', 'Day','Hour' columns.")
print("df1 head:\n", df1[['DATE', 'Year', 'Month', 'Day', 'Hour', 'Minute']].head())
print("df2 head:\n", df2[['DATE', 'Year', 'Month', 'Day', 'Hour', 'Minute']].head())
print("df3 head:\n", df3[['DATE', 'Year', 'Month', 'Day', 'Hour', 'Minute']].head())
```

```python
import numpy as np

# Convert 'HourlyDewPointTemperature' to numeric, coercing errors to NaN
df1['HourlyDewPointTemperature'] = pd.to_numeric(df1['HourlyDewPointTemperature'], errors='coerce')
df2['HourlyDewPointTemperature'] = pd.to_numeric(df2['HourlyDewPointTemperature'], errors='coerce')
df3['HourlyDewPointTemperature'] = pd.to_numeric(df3['HourlyDewPointTemperature'], errors='coerce')

# Group by Day and Hour, then calculate the mean of HourlyDewPointTemperature
df1_heatmap_data = df1.groupby(['Day', 'Hour'])['HourlyDewPointTemperature'].mean().unstack()
df2_heatmap_data = df2.groupby(['Day', 'Hour'])['HourlyDewPointTemperature'].mean().unstack()
df3_heatmap_data = df3.groupby(['Day', 'Hour'])['HourlyDewPointTemperature'].mean().unstack()

print("Heatmap data for df1 (first 5 rows and columns):")
print(df1_heatmap_data.iloc[:5, :5])
print("\nHeatmap data for df2 (first 5 rows and columns):")
print(df2_heatmap_data.iloc[:5, :5])
print("\nHeatmap data for df3 (first 5 rows and columns):")
print(df3_heatmap_data.iloc[:5, :5])
```

```python
import plotly.offline as pyo
import plotly.graph_objs as go
from plotly import tools

# Define common zmin and zmax for consistent color scaling across all heatmaps
common_zmin = min(df1_heatmap_data.min().min(), df2_heatmap_data.min().min(), df3_heatmap_data.min().min())
common_zmax = max(df1_heatmap_data.max().max(), df2_heatmap_data.max().max(), df3_heatmap_data.max().max())

# Create heatmap traces for each DataFrame
trace1 = go.Heatmap(
    x=df1_heatmap_data.columns,
    y=df1_heatmap_data.index,
    z=df1_heatmap_data.values,
    colorscale='Jet',
    zmin=common_zmin, zmax=common_zmax,
    name='Anchorage, AK'
)
trace2 = go.Heatmap(
    x=df2_heatmap_data.columns,
    y=df2_heatmap_data.index,
    z=df2_heatmap_data.values,
    colorscale='Jet',
    zmin=common_zmin, zmax=common_zmax,
    name='Buckley Releigh, WV'
)
trace3 = go.Heatmap(
    x=df3_heatmap_data.columns,
    y=df3_heatmap_data.index,
    z=df3_heatmap_data.values,
    colorscale='Jet',
    zmin=common_zmin, zmax=common_zmax,
    name='Seattle, WA'
)

# Create subplots
fig = tools.make_subplots(rows=1, cols=3,
    subplot_titles=('Anchorage, AK', 'Buckley Releigh, WV', 'Seattle, WA'),
    shared_yaxes=True,
    print_grid=False
)

# Append traces to the subplots
fig.append_trace(trace1, 1, 1)
fig.append_trace(trace2, 1, 2)
fig.append_trace(trace3, 1, 3)

# Update layout
fig['layout'].update(
    title='Daily and Hourly Dew Point Temperature Comparison',
    xaxis=dict(title='Hour'),
    yaxis=dict(title='Day of Month')
)

pyo.iplot(fig)
```

```python
import plotly.express as px
import pandas as pd

data = pd.read_csv('72793524234.csv')

# Convert 'DATE' column to datetime objects
data['DATE'] = pd.to_datetime(data['DATE'])

# Extract date components
data['Year'] = data['DATE'].dt.year
data['Month'] = data['DATE'].dt.month
data['Day'] = data['DATE'].dt.day
data['Hour'] = data['DATE'].dt.hour
data['Minute'] = data['DATE'].dt.minute

# Convert 'HourlyPrecipitation' to numeric, coercing errors to NaN
data['HourlyPrecipitation'] = pd.to_numeric(data['HourlyPrecipitation'], errors='coerce')

# Fill NaN values in 'HourlyPrecipitation' with 0, as missing precipitation data can be assumed as no precipitation
data['HourlyPrecipitation'] = data['HourlyPrecipitation'].fillna(0)

# Group by Month and sum the HourlyPrecipitation
monthly_precipitation = data.groupby('Month')['HourlyPrecipitation'].sum().reset_index()

# Create the bar chart with enhanced features
fig = px.bar(monthly_precipitation, x='Month', y='HourlyPrecipitation',
              title='Total Hourly Precipitation by Month in Seattle',
              labels={'HourlyPrecipitation': 'Total Precipitation (inches)', 'Month': 'Month'},
              color='HourlyPrecipitation', # Color bars by precipitation intensity
              color_continuous_scale='Blues', # Apply a blue color scale
              text='HourlyPrecipitation' # Add text labels to bars
             )

# Update the traces to format text labels to two decimal places and position them outside
fig.update_traces(texttemplate='%{text:.2f}', textposition='outside')

# Display the enhanced figure
fig.show()
```

This bar plot shows that summer months have less rainfall in Seattle compared to other months
